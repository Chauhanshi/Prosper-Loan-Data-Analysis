## Dataset Overview

> The data set contains 113,937 loans with 81 variables on each loan.

### Some important features:
>- LoanStatus : Current status of the loan like chargedoff, completed, defauted etc…
>- EstimatedEffectiveYield : Yield of lenders from borrowers minus the processing fee and late fines
>- BorrowerAPR : The Borrower’s Annual Percentage Rate (APR) for the loan.
>- BorrowerRate : The Borrower’s interest rate for this loan.
>- LoanOriginalAmount : Original amount of the loan
>- LoanOriginationQuarter : Quarter of the month when loan was originated
>- Term : Length of loan expressed in months

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
dfclean = pd.read_csv("loan_clean.csv")
```

```python
dfclean.IncomeRange = dfclean.IncomeRange.str.replace('Not employed','$0')

# converting Income Range into ordered categorical types
ordinal_var_dict = {'IncomeRange': ['Not displayed','$0','$1-24,999','$25,000-49,999','$50,000-74,999','$75,000-99,999',
                                   '$100,000+']}


for var in ordinal_var_dict:
    pd_ver = pd.__version__.split(".")
    if (int(pd_ver[0]) > 0) or (int(pd_ver[1]) >= 21): # v0.21 or later
        ordered_var = pd.api.types.CategoricalDtype(ordered = True,
                                                    categories = ordinal_var_dict[var])
        dfclean[var] = dfclean[var].astype(ordered_var)
    else: # pre-v0.21
        dfclean[var] = dfclean[var].astype('category', ordered = True,
                                             categories = ordinal_var_dict[var])
        
df1 = dfclean.query("LoanOriginationQuarter in ['Q4 2013','Q1 2014','Q3 2013']")

#creating a different column for quarter and year
dfclean['quarter'] = dfclean.LoanOriginationQuarter.apply(lambda x: x.split()[0])
dfclean['year'] = dfclean.LoanOriginationQuarter.apply(lambda x: x.split()[1])

#creating a different column for quarter and year
df1['quarter'] = df1.LoanOriginationQuarter.apply(lambda x: x.split()[0])
df1['year'] = df1.LoanOriginationQuarter.apply(lambda x: x.split()[1])
df1.year = pd.to_datetime(df1.year)
```

## Which loan term are given most and when most of the loan are given? 

>- The vizualization shows that most people opt for mid-term loan which is 36 months term loan. 
>- Data also shows most of the loan were taking during July 2013 to March 2014. 

```python
#same plot with some difference
plt.figure(figsize = [14.70, 8.27])

# left plot:
plt.subplot(1, 2, 1)
sorted_counts = dfclean.Term.value_counts()
plt.pie(sorted_counts, labels = ['36 month','60 month','12 month'], startangle = 90,
        counterclock = False, wedgeprops = {'width' : 0.4});
plt.axis('square');
plt.legend();
plt.title('Count of Loan Term in Months',size =15);


# right plot: 
plt.subplot(1, 2, 2)
name = [""] * len(dfclean.LoanOriginationQuarter.value_counts())
#name[:] = np.nan
c = dfclean.LoanOriginationQuarter.value_counts().index[0]
d = dfclean.LoanOriginationQuarter.value_counts().index[1]
e = dfclean.LoanOriginationQuarter.value_counts().index[2]
name[0:3] = c,d,e
sorted_counts = dfclean.LoanOriginationQuarter.value_counts()
plt.pie(sorted_counts, labels = name, startangle = 90,
        counterclock = False, wedgeprops = {'width' : 0.4});
plt.axis('square');
plt.title("Count of Loan distribution per quarter", size=15);
```

## Which income group takes most loans and  how much?

>- Most Loans are taken by mid-range income group which is 25,000USD - 74,999USD.
>- As the income range increases, the avg loan amount taken per group also increases. 
>- Higher the income range group, Standard deviation in the avg loan amount increases, this means, higher income group customer can opt for small loans as well as higher amount loans. 

```python
plt.figure(figsize = [14.70, 8.27])

# left plot:
plt.subplot(1, 2, 1)
#bar chart for Income range 
base_color = sns.cubehelix_palette(dfclean.IncomeRange.nunique(),reverse=False) #(sns.cubehelix_palette(7))
sns.countplot(x = dfclean.IncomeRange,order=ordinal_var_dict['IncomeRange'],palette=base_color);
plt.title("Count of loan given as per Income Range",size = 12);
plt.xticks(rotation=45);


# right plot: 
plt.subplot(1, 2, 2)

sns.pointplot(data = dfclean, x = dfclean.IncomeRange, y = dfclean.LoanOriginalAmount , ci = 'sd',linestyles=""); 
plt.xticks(rotation=45);
plt.ylabel("average loan amount");
plt.title("Avg loan amount and standad deviation as per income group");
```

## What is trend for BorrowerAPR per quarter from 2006 to 2014?

> We found that average APR was increasing till 2011 Q3 and then it started to decline. Average APR was dropped tremendously over 3 years which may be the reason why after 2013 Q3 so many loans are taken by the customers. 

```python
plt.figure(figsize = [14.70, 8.27])

sns.barplot(data = dfclean, x = 'year', y = dfclean.BorrowerAPR ,hue = dfclean.quarter,hue_order=['Q1','Q2','Q3','Q4'],dodge=True); 
plt.xticks();
plt.ylabel("average of APR",size= 10)
plt.title('Avg BorrowerAPR trend per quarter from 2006 to 2014',size = 15);
```

## What is trend for Loan Amount per quarter from 2006 to 2014?

> Data Shows Avg Loan Amount is continuously increasing from 2009 Q1. Highest avg loan amount is again during the period of July 2013 to March 2014. This may be because the Borrower's APR has declined tremendously during this period. It also shows more and more people are taking loans from 2009. Avg Loan amount given is approximatly tripped from 2009 to 2014.

```python
plt.figure(figsize = [14.70, 8.27])

sns.barplot(data = dfclean, x = 'year', y = dfclean.LoanOriginalAmount ,hue = 'quarter',hue_order=['Q1','Q2','Q3','Q4'],dodge=True); 
plt.xticks();
plt.ylabel("average of Amount",size= 10)
plt.title('Avg Loan Amount trend per quarter from 2006 to 2014',size = 15);
```

## Conclusion

>To conclude, we saw that most loan given are mid term loan with balance risk to repay in 36 months. Usually mid-range income groups takes most loans. We observed a strong negative relation between Borrower's APR and Loan Amount. Borrower's APR declined from 2011 which resulted in people taking more loan amount with lower interest rate. The trend of loan taking is showing an upward trend along with the avg loan amount. There are other factors also affect the decision to take loan because longer term means more amount you will have to pay as interest. Similarly, lower term loans have high risk if you are unable to repay the loan amount. Further analysis is required for finding why most people are opting for higher amount loans.
